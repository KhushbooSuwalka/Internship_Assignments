```python
# QUES 3
t = (10, 20, 30, 40)
element = int(input("Enter an element :"))
print("Element : ",element)

if element in t:
    print("Element exists in tuple")
else:
    print("Element does not exist")
```

```python
# QUES 4
t = (12, 45, 7, 23, 89, 15)
total = 0
highest = t[0]
lowest = t[0]

for num in t:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Total =", total)
print("Highest =", highest)
print("Lowest =", lowest)
```

```python
# QUES 5
n = (3, 14, 7, 22, 9, 41, 18, 5)
result = ()

for i in n:
    if i > 10:
        result += (i,)
print("Filtered Tuple:", result)
```

```python
# QUES 6
s = {"cat", "dog", "bird", "fish"}
count = 0

for item in s:
    count += 1
print("Number of elements:", count)
```
