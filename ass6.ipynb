```python
# Ques 3 
data = {
    "ID": [1, 2, 3, 4],
    "Name": ["Khushboo", "Riya", "Aman", "Rahul"],
    "Marks": [85, 92, 76, 60]
}
df = pd.DataFrame(data)
print(df)

# printing each row of the DataFrame
print("\nShowing the data row by row:")
for i in range(len(df)):
    print(df.iloc[i]["ID"],df.iloc[i]["Name"], df.iloc[i]["Marks"])

# printing the first 3 rows of the DataFrame
print("\nShowing the first 3 rows of the DataFrame:")
print(df.loc[0:2]) 
```

```python
print("Students with Marks > 80:")
print(df[df["Marks"] > 80])
```
