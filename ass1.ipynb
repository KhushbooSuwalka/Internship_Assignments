```python
# Program to calculate total marks and percentage of a student
name=input("Enter the name of the student: ")
stu_class = int(input("Enter the class of the student: "))
total_marks=0
for i in range(1,6):
    marks=float(input(f"Enter the marks of subject {i}: "))
    total_marks+=marks
percentage=total_marks/5
print(f"Name: {name}")
print(f"Class: {stu_class}")
print("Total marks:", total_marks)
print("Percentage:", percentage)
```

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")
result = str1 + " " + str2
print("\nConcatenated String:", result)

print("\n--- String Methods Output ---")
print("lower()      :", result.lower())
print("upper()      :", result.upper())
print("title()      :", result.title())
print("swapcase()   :", result.swapcase())
print("capitalize() :", result.capitalize())
print("casefold()   :", result.casefold())
print("center(50)   :", result.center(50, '*'))

word = input("\nEnter word/character to count: ")
print("count()      :", result.count(word))

end_word = input("Enter ending word to check: ")
print("endswith()   :", result.endswith(end_word))

find_word = input("Enter word to find: ")
print("find()       :", result.find(find_word))

print("isalnum()    :", result.isalnum())
print("isdigit()    :", result.isdigit())
print("isnumeric()  :", result.isnumeric())
print("isspace()    :", result.isspace())

old = input("Enter word to replace: ")
new = input("Enter new word: ")
print("replace()    :", result.replace(old, new))
```

```python
# Practice of Assignment Operators in Python
a = 20
print("Initial value of a =", a)
a += 5
print("After +=  :", a)
a -= 3
print("After -=  :", a)
a *= 2
print("After *=  :", a)
a /= 4
print("After /=  :", a)
a %= 3
print("After %=  :", a)
a **= 2
print("After **= :", a)
a //= 2
print("After //= :", a)
```

```python
name=input("Enter the name of the student: ")
stu_class = int(input("Enter the class of the student: "))
total_marks=0
for i in range(1,6):
    marks=float(input(f"Enter the marks of subject {i}: "))
    total_marks+=marks
percentage=total_marks/5
if percentage >=60:
    grade = "A"
elif percentage >=50 and percentage <60:
    grade = "B"
elif percentage >=40 and percentage <50:
    grade = "C" 
elif percentage >=33 and percentage <40:
    grade = "D"
else:
    grade = "Fail"
print("Student Details:")  
print(f"Name: {name}")
print(f"Class: {stu_class}")
print("Total marks:", total_marks)
print("Percentage:", percentage)
print("Grade:", grade)
```
