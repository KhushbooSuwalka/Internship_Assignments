```python
# QUES 1
import numpy as np
arr = np.array([[6, -8, 73, -110],
                [np.nan, -8, 0, 94]])
print("Original Array:")
print(arr)

arr = np.nan_to_num(arr, nan=0)
print("\nAfter Replacing NaN with 0:")
print(arr)

arr_transpose = arr.T
print("\nAfter Interchanging Rows and Columns:")
print(arr_transpose)

print("\nFirst 3 Rows and 3 Columns:")
print(arr_transpose[:3, :3])
```

```python
# QUES 2
arr = np.arange(24).reshape(2, 3, 4)
print("Original Shape:", arr.shape)

# Move axis 0 to position 2
new_arr = np.moveaxis(arr, 0, 2)
print("New Shape:", new_arr.shape)
print("\nArray after Moving Axes:")
print(new_arr)
```

```python
# QUES 3
import pandas as pd
data = {
    "Math": [90, 80, np.nan, 70],
    "Science": [85, np.nan, 75, 95]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df.fillna(df.mean(), inplace=True)
print("\nDataFrame after Replacing NaN with Column Average:")
print(df)
```

```python
# QUES 4
arr = np.array([10, -5, 20, -8, 30, -2])
print("Original Array:")
print(arr)

arr[arr < 0] = 0
print("\nArray after replacing negative values with 0:")
print(arr)
```

```python
# QUES 5
arr1 = np.array([3, 4]) 
arr2 = np.array([1, 0])
print("First Array:", arr1)
print("Second Array:", arr2)

avg = (arr1 + arr2) / 2
print("\nAverage of Arrays:")
print(avg)
```

```python
arr1 = np.array([[10, 20],
                 [30, 40]])
arr2 = np.array([[50, 60],
                 [70, 80]])
print("First 2D Array:")
print(arr1)
print("\nSecond 2D Array:")
print(arr2)

avg = (arr1 + arr2) / 2
print("\nAverage of Arrays:")
print(avg)

from statistics import mode
combined = np.concatenate((arr1, arr2))

print("Mean:", np.mean(combined))
print("Median:", np.median(combined))
print("Mode:", mode(combined.flatten()))
```

```python
# QUES 6
A = np.array([
    [1, -2, 3],
    [-1, 3, -1],
    [2, -5, 5]
])
B = np.array([9, -6, 17])

A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("Solution:")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

```python
# QUES 7
import matplotlib.pyplot as plt
subjects = ["Math", "Python", "DBMS", "OS", "DSA"]

sem1 = [75, 80, 70, 78, 82]
sem2 = [85, 88, 80, 84, 90]

plt.plot(subjects, sem1,
         marker='o',
         linewidth=2,
         label='Semester 1')
plt.plot(subjects, sem2,
         marker='s',
         linewidth=2,
         label='Semester 2')

plt.title("Semester Result Comparison")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(True)
plt.legend()

plt.show()
```
