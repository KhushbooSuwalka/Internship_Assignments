```python
# QUES 1
# Convert a Series of date-strings to a TimeSeries
import pandas as pd
dates = pd.Series([
    "2026-05-20",
    "2026-05-21",
    "2026-05-22"
])
print("Original Series:")
print(dates)

time_series = pd.to_datetime(dates)
print("\nConverted TimeSeries:")
print(time_series)
```

```python
#QUES 2
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Khushboo", "Riya", "Aman", "Rahul"]
})
df2 = pd.DataFrame({
    "ID": [2, 3, 4, 5],
    "Marks": [85, 90, 78, 88]
})
print("DataFrame df1:")
print(df1)
print("\nDataFrame df2:")
print(df2)
```

```python
inner_merge = pd.merge(df1, df2, on="ID", how="inner")
print("Inner Merge:")
print(inner_merge)
```

```python
left_join = df1.merge(df2, on="ID", how="left")
print("Left Join:")
print(left_join)
# Explanation : If a matching ID is not found in df2, Pandas fills the missing value with NaN.
```

```python
right_join = df1.merge(df2, on="ID", how="right")
print("Right Join:")
print(right_join)
# Explanation : If a matching ID is not found in df1, Pandas fills the missing value with NaN.
```

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Khushboo", "Riya", "Aman", "Rahul"]
},index=["a", "b", "c", "e"])
df2 = pd.DataFrame({
    "ID": [2, 3, 4, 5],
    "Marks": [85, 90, 78, 88]
},index=["a", "b", "c", "d"])

print("Indexed based join:")
res=df1.join(df2,lsuffix="left",rsuffix="right",how="inner") 
print(res)
```

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Khushboo", "Riya", "Aman", "Rahul"]
})
df2 = pd.DataFrame({
    "ID": [2, 3, 4, 5],
    "Marks": [85, 90, 78, 88]
})
right_join = pd.merge(df1,df2, on="ID", how="right")
print("Right Join using merge:")
print(right_join)

df1.index = ["a", "b", "c", "e"]
df2.index = ["a", "b", "c", "d"]
print("Right Join using join:")
res = df1.join(df2, lsuffix="left", rsuffix="right",how="right")
print(res)

# Comparing the results :
# Both give same result but merge can reduce the number of rows and columns in the result if there are duplicate keys in the join columns, while join will keep all rows from both DataFrames.
```

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Dept": ["CS", "IT", "CS"],
    "Marks": [80, 90, 85]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Dept": ["CS", "IT", "ECE"],
    "Grade": ["A", "A+", "B"]
})

multi_merge = pd.merge(df1, df2, on=["ID", "Dept"], how="inner")
print("Merge with Multiple Keys:")
print(multi_merge)
```

```python
#QUES 3
df1 = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["Khushboo", "Riya"]
})
df2 = pd.DataFrame({
    "ID": [3, 4],
    "Name": ["Aman", "Rahul"]
})
df3 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Marks": [85, 90, 78, 88]
})

print("DataFrame df1:")
print(df1)
print("\nDataFrame df2:")
print(df2)
print("\nDataFrame df3:")
print(df3)
```

```python
res=pd.concat([df1,df2],ignore_index=True) 
print("\nConcatenating along rows:")
print(res)

final_res=res.merge(df3, on="ID", how="inner")
print("\nMerging with df3 :")
print(final_res)
```

```python
print("Difference between join() and merge():")
print("""
pd.merge()
--------
1. Used to combine DataFrames using common columns.
2. Syntax: pd.merge(df1, df2, on='column')
3. More flexible.
4. Default join column is specified using 'on='.
5. merge() can perform: inner join,left join,right join,outer join

df.join()
-------
1. Used to combine DataFrames using index.
2. Simpler for index-based joining.
3. Faster for joining on index.
4. Syntax: df1.join(df2)
5. lsuffix and rsuffix parameters can be used to handle overlapping column names.
""")
```
